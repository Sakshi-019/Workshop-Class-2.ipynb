```python
class track:
    def __init__(self, coor1, coor2):
        self.coor1 = coor1
        self.coor2 = coor2
        
    def distance(self):
        x1,y1 = self.coor1
        x2,y2 = self.coor2
        return ((x2 - x1)*2 - (y2 - y1)**2)*0.5
    
    def slope(self):
        x1,y1 = self.coor1
        x2,y2 = self.coor2
        return (y2 - y1)/(x2 - x1)
```

```python
c1 = (4, 9)
c2 = (9, 12)
  
```

```python
find = track(c1, c2)
```

```python
   find.distance()
```

```python
 find.slope()
```

```python
class Account:
    def __init__(self,owner,balance=0):
        self.owner = owner
        self.balance = balance
        
    def _str_(self):
        return f'Account owner:     {self.owner}\nAccount balance:   ${self.balance}'
    
    
    def deposit(self,dep_amt):
        self.balance += dep_amt
        print("Deposit Accepted")
        
        
    def withdraw(self,wd_amt):
        if self.balance >= wd_amt:
            self.balance -= wd_amt
            print('Withdrawal Accepted')
            
        else:
            print('Funds Unavailaable')
```

```python
acc1 = Account('Saurabh',20000)
 
```
